# handle null values

```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

# rename the colums, make all lower case & use underscore instead of space

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.columns
```

# derive new column for discount

```python
df['discount'] = df['list_price']*df['discount_percent']/100
```

```python
df
```

# derive new column sale_price

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df
```

# derive new column profit

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df
```

# convert order_date column from object type to datetime

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

# Establish connection to MySQL

```python
from sqlalchemy import create_engine
```

```python
# MySQL connection details
username = 'root'
password = 'root'
host = 'localhost'
database = 'retails_orders_db'

# Create a connection engine
engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}/{database}')

```

```python
# Load DataFrame into MySQL
df.to_sql(name='df_orders', con=engine, if_exists='append', index=False)
```
